Case Study – Student Marks Analysis
Instructions:

Create a list named marks by getting 5 subject marks from the user using map() and list()

Perform calculations:

• Find total marks using sum()

• Find average marks using len()

• Calculate percentage assuming each subject is out of 100

```python
Marks = list(map(int,input("Enter the Marks: ").split(",")))
Marks
```

```python
#Total Marks

Total_Marks = sum(Marks)
print("Total Marks Scored:",Total_Marks)
```

```python
#Avg Marks
average_Marks = Total_Marks/len(Marks)
print("Average Mark Scored:", average_Marks)
```

```python
#Percentage of marks
max_Marks = len(Marks)*100
percentage = (Total_Marks/max_Marks)*100
print("Percentage:",percentage,"%")
```

```python
print("Maximum Marks:",max(Marks))
print("Minimum Marks:",min(Marks))
```
